```python
train=(transactions.groupby(['shop_id','item_id','date_block_num']).agg({'item_cnt_day': 'sum',})
.reset_index()
.rename(columns={'item_cnt_day':'item_cnt_month'})
)

train['item_cnt_month']=train['item_cnt_month'].clip(0.0,20.0)

oct_2015 = train[['shop_id','item_id','item_cnt_month']][train['date_block_num']==33]

test = df5.merge(oct_2015,on=['shop_id','item_id'],how='left')

test['item_cnt_month']=test['item_cnt_month'].fillna(0)

test.sort_values('ID',inplace=True)
```

```python
submission = test.loc[:,['ID']]
submission['item_cnt_month'] = test['item_cnt_month']
# submission.sort_values('ID') -> should be 214200,2
print(submission.shape[0])
submission.to_csv('prev_value_benchmark.csv', index=False)
```
